```python
# for manipulations
import numpy as np
import pandas as pd

# for data visualizations
import matplotlib.pyplot as plt
import seaborn as sns

# for interactivity
from ipywidgets import interact
```

```python
# lets read the dataset
data = pd.read_csv('denguefever.csv')
```

```python
# lets check the shape of the dataset
print("Shape of the Dataset :", data.shape)
```

```python
# Lets check the head of the dataset
data.head()
```

```python
# lets check if there is any missing value present in the dataset
data.isnull().sum()
```

```python
# lets check the crops present in this dataset
data['Name'].value_counts()
```

```python
# lets check the summary for all symptoms
# Mean

print("----- Mean of the Dataset -----\n")
print(f"Age Who have suffered from Dengue Fever: {data['Age'].mean():.2f}")
print(f"Duration of Fever: {data['Duration of fever'].mean():.2f}")
print(f"Haemoglobin Level: {data['Haemoglobin Level'].mean():.2f}")
print(f"PCV Count: {data['PCV count'].mean():.2f}")
print(f"Platelets Count: {data['Platelets Count'].mean():.2f}")
print(f"Leukocytes Count: {data['Leukocytes Count'].mean():.2f}")
print(f"RBC Count: {data['RBC Count'].mean():.2f}")
```

```python
#Median

print("----- Median of the Dataset -----\n")
print(f"Age Who have suffered from Dengue Fever: {data['Age'].median():.2f}")
print(f"Duration of Fever: {data['Duration of fever'].median():.2f}")
print(f"Haemoglobin Level: {data['Haemoglobin Level'].median():.2f}")
print(f"PCV Count: {data['PCV count'].median():.2f}")
print(f"Platelets Count: {data['Platelets Count'].median():.2f}")
print(f"Leukocytes Count: {data['Leukocytes Count'].median():.2f}")
print(f"RBC Count: {data['RBC Count'].median():.2f}")
```

```python
# Covariance

print("----- Covariance of the Dataset -----\n")
print("Duration of Fever: {:.2f}".format(data['Duration of fever'].cov(data['Age'])))
print("Haemoglobin Level: {:.2f}".format(data['Haemoglobin Level'].cov(data['Age'])))
print("PCV Count: {:.2f}".format(data['PCV count'].cov(data['Age'])))
print("Platelets Count: {:.2f}".format(data['Platelets Count'].cov(data['Age'])))
print("Leukocytes Count: {:.2f}".format(data['Leukocytes Count'].cov(data['Age'])))
print("RBC Count: {:.2f}".format(data['Leukocytes Count'].cov(data['Age'])))
```

```python
# Standard Deviaton

print("----- Standard Deviation of the Dataset -----\n")
print("Duration of Fever: {:.2f}".format(data['Duration of fever'].std()))
print("Haemoglobin Level: {:.2f}".format(data['Haemoglobin Level'].std()))
print("PCV Count: {:.2f}".format(data['PCV count'].std()))
print("Platelets Count: {:.2f}".format(data['Platelets Count'].std()))
print("Leukocytes Count: {:.2f}".format(data['Leukocytes Count'].std()))
print("RBC Count: {:.2f}".format(data['RBC Count'].std()))
```

```python
#Correlation

print("----- Correlation between Age And other attributes of the Dataset -----\n")
print("Duration of Fever and Age: {:.2f}".format(data['Duration of fever'].corr(data['Age'])))
print("Haemoglobin Level and Age: {:.2f}".format(data['Haemoglobin Level'].corr(data['Age'])))
print("PCV Count and Age: {:.2f}".format(data['PCV count'].corr(data['Age'])))
print("Platelets Count and Age: {:.2f}".format(data['Platelets Count'].corr(data['Age'])))
print("Leukocytes Count and Age: {:.2f}".format(data['Leukocytes Count'].corr(data['Age'])))
print("RBC Count and Age: {:.2f}".format(data['RBC Count'].corr(data['Age'])))
```

```python
# Lets check the Summary Statistics for each of the Patient

@interact
def summary(patient=list(data['Name'].unique())):
    x = data[data['Name'] == patient].iloc[0]
    print("--------------------------------------------------------")
    print("------ Stats of the Patient -----\n")
    print("Duration of Fever :", x['Duration of fever'])
    print("Haemoglobin Level :", x['Haemoglobin Level'])
    print("PCV count :", x['PCV count'])
    print("Platelets Count :", x['Platelets Count'])
    print("Leukocytes Count :", x['Leukocytes Count'])
    print("RBC Count :", x['RBC Count'])
    
```

```python
# Average level according to attribute as input
@interact
def compare(Condition = ['Duration of fever', 'Haemoglobin Level', 'PCV count', 'Platelets Count', 'Leukocytes Count', 'RBC Count']):
    print("--- Patients whose",Condition,"is greater than average ---", '\n')
    print(data[data [Condition] > data [Condition].mean()]['Name'].unique())
    print("\n-------------------------------------------------------------------\n")
    print("--- Patients whose",Condition,"is less than average ---", '\n')
    print(data[data [Condition] <= data[Condition].mean()]['Name'].unique())
```

```python
#lets Understand the condition of the patient

print("---------------- BAD CONDITION ----------------\n")
print(data[(data['Platelets Count'] > 200000)]['Name'].unique())
print("-------------------------------------------------------")
print("\n--------------- WORSE CONDITION ---------------\n")
print(data[(data['Platelets Count'] > 100000) & (data['Platelets Count'] < 200000)]['Name'].unique())
print("-------------------------------------------------------")
print("\n--------------- WORST CONDITION ---------------")
print(data[(data['Platelets Count'] < 100000) & (data['Platelets Count'] > 0)]['Name'].unique())
```
